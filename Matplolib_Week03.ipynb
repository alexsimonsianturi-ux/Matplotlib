```python
#Matplohib weeek 03
#Nama : Alex simon sianturi
#Nrp : 2473036
#Pengenalan pyplot
%matplotlib inline
import matplotib
import matplotib.pyplot as plt
import numpy as np

print(matplotib.__version__)
print(np.__version__)
```

```python
#Simple Line Plot
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2*np.pi*x)
s
```

```python
fig, ax= plt.subplots() #00 Style
ax.plot(x,s)
ax.set(xlable='nilai x',
        ylabel= 'nilai sin(x)',
        title='Visualisasi nilai sin')
ax.grid()
plt.show()
```

```python
#Multiple Subplots
x1= np.linspace(0.0, 5.0, 100)
x2= np.linespace(0.0, 2.0, 100)

y1=np.cos(2*np.pi*x1)
y2=np.cos(2*np.pi*x2)
x1   
```

```python
#Multiple Subplots dengan OO Style
fig, (ax1, ax2)= plt.subplots(2,1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')
ax2.set_xlabel('Nilai x')
plt.show()
```

```python
#Multiple subplots dengan pyplot style
plt.subplot(2,1,1)
plt.plot(x1,y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai ')
plt.xlabel('Nilai ')
plt.show
```

```python
#04 Histogram
mu, sigma = 100, 15 # nilai mean dan nilai standard devation
x= mu + sigma * np.random.radn(10000) # normal distribution 
x
x.shape
```

```python
# Histrogram dengan pyplot Style
plt.hist(x,
        bins=50
        facecolor='g',
        alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.tittle('Contoh Histogram')

plt.text(45, 500, r'')
plt.grid()
plt.show()
```

```python
#Histogram dengan OO Style
fig, ax = plt.subplots()
plt.hist(x,
        bins=50
        facecolor='r',
        alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.tittle('Contoh Histogram')

plt.text(45, 500, r'')
plt.grid()

plt.show()
```
